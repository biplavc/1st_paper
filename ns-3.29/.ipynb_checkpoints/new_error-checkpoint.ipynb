```python
def calc_error(c): # c is the part of the ns3 dataframe for a fixed tx-rx pair
    
    nrows = c.shape[0]
    if nrows==1:
        return c["Error(m)"].values[0]
    else:
        loops = nrows - 1 # as loops will be between 2 BSMs, so no of loops = no of BSMs - 1
        sender = c.loc[0, "TxId"] # sender will be the same throughout, so take the sender value from any row
        receiver = c.loc[0, "RxId"] # same as above
        mask = (sumo_df['VID'] == sender) # extract the senders location info from global sumo file
        df_actual = sumo_df[mask]
#         print(df_actual) # working OK
    
    
#         for(i = 0, i < loops, i++): # no of times the loop will see for missing BSMs = take 2 BSMs at once
        for i in range(loops):
            error = 0 # will hold the total error
            current_bsm = c.loc[i]
            next_bsm =  c.loc[i+1]
            slots = int(next_bsm['RecTime'] - current_bsm['RecTime'] - 1) # no of times the next loop will run
            current_slot = int(current_bsm["RecTime"])
            x_pos_BSM = c.loc[i, "TxHeadPos(X)"]
            y_pos_BSM = c.loc[i, "TxHeadPos(Y)"]
            x_speed_BSM = c.loc[i, "Tx_HeadVelX"]
            y_speed_BSM = c.loc[i, "Tx_HeadVelY"]
            error = error + c.loc[i, "Error(m)"]
            print("error at reception slot", current_slot," is", error)
#             print("x_pos_BSM = ", x_pos_BSM, "x_speed_BSM = ", x_speed_BSM, "error = ", error ) 
            # error has been initialized to 0, so here add the error at 
            # moments of reception, in the next loop, add the error from tracking

#             for j=1, j<=slots, j++):
            for j in range(slots):
    ## predcition = last known location + last known velocity * no of time slots
                print("current slot for prediction = ", current_slot+j+1) 
                x_pos_predicted = x_pos_BSM + (x_speed_BSM * (j+1))*(0.001) # as slots are in msec and speed in m/s
                y_pos_predicted = y_pos_BSM * (y_speed_BSM * (j+1))*(0.001)
                mask = df_actual["Time"] == (current_slot + (j+1))*(0.001)
                df_row = df_actual[mask] 
                # row of sumo file at the ongoing slot
                x_pos_actual = df_row["Pos(X)"].values[0]
                y_pos_actual = df_row["Pos(Y)"].values[0]
                print("x actual=", x_pos_actual," y actual=",y_pos_actual," x pred=",x_pos_predicted, " y pred =", y_pos_predicted)
                error_x = abs(x_pos_predicted - x_pos_actual)
                error_y = abs(y_pos_predicted - y_pos_actual)
                print("x error = ", error_x, ", y error = ", error_y)
                slot_error = sqrt(error_x*error_x + error_y*error_y)
                print("error at prediction slot ", current_slot+j+1, " is ", slot_error)
                error = error + slot_error
        return error
```

```python
for file in ns3_files:
    b = pd.read_csv(file, delim_whitespace = True)
    b = b.sort_values(['RecTime'])
    b = b.reset_index(drop=True)
    m = b['RxId'].nunique() # m is number of receivers
    # throughput block starts
    overall_duration = (b['RecTime'].max() - b['RecTime'].min())/1000 # milliseconds to seconds
    
    ## in throughput case, we work on the whole file so 1 pair or 1 packet exchanged cases do not apply.
    
    packets = b.shape[0] # no of rows = no of packets received
    throughput = (packets/overall_duration)*8*0.32 # in Kbps as each packet 320 Byte
    throughput = throughput/1000 # Mbps
    average_throughput = throughput/m
    total_system_throughput.append(throughput) # one entry per file
    average_system_throughput.append(average_throughput) # one entry per file
    
    collisions = 0
    
    receivers = b['RxId'].unique()
    
    average_ages = [] # hold age for every pair in a file
    average_errors = [] # hold error for every pair in a file
    
#     average_collisions = []
    for i in range(len(receivers)):
        receiver = receivers[i] # for every receiver
#         receiver_velocity = float(b2.Velocity[b2['ID']==receiver])
        senders = b[b['RxId'] == receiver].TxId.unique() 
#             print(receiver, senders)
        for j in range(len(senders)): 
            sender = senders[j]
            mask = (b['RxId'] == receiver) & (b['TxId'] == sender) # extract out the rx-tx pair
            c = b[mask]
            c = c.reset_index(drop=True)          
            y = c.shape[0]  # gives number of row count
            last_pack_rec = np.max(c["RecTime"]) # milliseconds
            last_known_x_pos = (c[c['RecTime']==last_pack_rec])["TxHeadPos(X)"]
            last_known_y_pos = (c[c['RecTime']==last_pack_rec])["TxHeadPos(Y)"]
            # place from where the last packet was received
            # print("last_pack_rec = ", last_pack_rec)
            recent = 0 # will keep track for any non-informative packets, time
            for z in range(y): # for all the rows
                if c.loc[z,"GenTime"] >= recent:
                    recent = c.loc[z,"GenTime"]
                else: # will work when LCFS used
                    c.loc[z,"GenTime"] = np.NaN
                    c.loc[z,"Delay"] = np.NaN
                    c.loc[z,"RecTime"] = np.NaN
                    c.loc[z,"Error(m)"] = np.NaN
                    
            # the df has been modified to make the non-informative packet contents as NaN
            gentime = np.array(c['GenTime'])
            rectime = np.array(c['RecTime'])
            delay = np.array(c['Delay'])
            error = np.array(c['Error(m)'])
            xpos_header = np.array(c['TxHeadPos(X)'])
            ypos_header = np.array(c['TxHeadPos(Y)'])
            xpos_true = np.array(c['TxTruePos(X)'])
            ypos_true = np.array(c['TxTruePos(Y)'])
            xvel_header = np.array(c['Tx_HeadVelX'])
            yvel_header = np.array(c['Tx_HeadVelY'])
            
            # pos_error = error(gentime, rectime, error, xpos_header, ypos_header, xpos_true, ypos_true, xvel_header, yvel_header)
            print("error calculation will be done for sender ",sender, " and receiver ", receiver)
            avg_error = calc_error(c) # calc_error is the function
            # this will give the whole error for that pair
            # pos_error should return a value of error
            
#             avg_error = np.sum(pos_error)/overall_duration # errors for single pair
            average_errors.append(avg_error) # average_errors will hold the error for every pair in a file
        
    if m == 1: # m is number of receivers, i.e. a single pair
        # average error will hold the average of all the errors for every pair = avg of average_errors
        average_error = avg_error
        
    else:
        average_error = np.average(average_errors)
        
    
    print("average_error = ", average_error)
```
